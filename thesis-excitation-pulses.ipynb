```python
%matplotlib notebook
from utils.converter import *
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('support-files/qubit-01.mplstyle')
```

```python
FILENAME = "data/excitation-pulses/monitoring-excitation-pulse_direct-from-keysight_1MHz40ns_CHA-gausss_CHB-pulse_48MHz-filter.txt"
DATA = np.loadtxt(FILENAME)

fig, ax = plt.subplots(1,1,figsize=(3,2))

ax.plot(DATA[0],color="C5", linewidth=0.8,marker="o",label="Gaussian (40ns)")
ax.plot(DATA[1],color="C6", linewidth=0.8,marker="o",label="Square (40ns)")

ax.set_ylabel("Codes")
ax.set_xlabel("Time (ns)")

ax.legend(ncol=2)
plt.tight_layout()
plt.savefig(FILENAME.replace(".txt",".svg"))
```
